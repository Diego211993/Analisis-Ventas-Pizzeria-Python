```python
nombre_archivo = 'pizza_sales.csv'
```

```python
df_pizza = pd.read_csv(nombre_archivo)
df_pizza.head()
```

```python
df_pizza.info()
```

```python
df_pizza['pizza_size'].value_counts()
```

```python
df_pizza.isnull().sum()
```

```python
df_pizza.describe()
```

```python
total_ventas = df_pizza['total_price'].sum()
print(f"La venta total fue de: {total_ventas}")
```

```python
ticket_promedio = df_pizza['unit_price'].mean()
print(ticket_promedio)
```

```python
ventas_por_categoria = df_pizza.groupby('pizza_category')['quantity'].sum().sort_values(ascending=False)
print(ventas_por_categoria)
```

```python
top_pizzas = df_pizza.groupby('pizza_name')['quantity'].sum().nlargest(5)
print(top_pizzas)
```

```python
ventas_por_tamano = df_pizza.groupby('pizza_size')['quantity'].sum()
print(ventas_por_tamano)
```

```python
df_pizza['order_date'] = pd.to_datetime(df_pizza['order_date'], format='mixed')
```

```python
df_pizza.info()
```

```python
df_pizza["dia_semana"] = df_pizza['order_date'].dt.day_name()
df_pizza.head()
```

```python
df_pizza['mes'] = df_pizza['order_date'].dt.month_name()
print(df_pizza['mes'].value_counts())
```

```python
df_pizza['dia_semana'].value_counts().plot(kind='bar', title='Ventas por da de la semana', color = "blue")
```

```python
ingresos_categoria = df_pizza.groupby('pizza_category')['total_price'].sum().sort_values()
print(ingresos_categoria)
```

```python
ingresos_categoria.plot(kind='barh', color='green', title='Ingresos Totales por Categora ($)')
```

```python
aov = df_pizza.groupby('order_id')['total_price'].sum().mean()
print(f"el gasto promedio por orden es de: ${aov:2f}")
```

```python
total_ventas = df_pizza['total_price'].sum()
print(f"La venta total fue de: {total_ventas}")
```

```python
total_ordenes = df_pizza['order_id'].nunique()
print(f"El total de ordenes fue de: {total_ordenes}")
```

```python
df_pizza['hora'] = pd.to_datetime(df_pizza['order_time'], format='%H:%M:%S').dt.hour
df_pizza.groupby('hora')['quantity'].sum().plot(kind='line', title="Demanda por hora")
```

```python
import matplotlib.pyplot as plt
```

```python
ventas_por_hora = df_pizza.groupby('hora')['quantity'].sum()
```

```python
plt.figure(figsize=(12, 6))
ventas_por_hora.plot(kind='line', marker='o', color='darkorange', linewidth=2)
plt.title('Demanda de Pizzas por Hora del Da', fontsize=15)
plt.xlabel('Hora (Formato 24h)', fontsize=12)
plt.ylabel('Cantidad de Pizzas Vendidas', fontsize=12)
plt.xticks(range(9, 24)) # Ajustamos el eje X para ver todas las horas
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import seaborn as sns
```

```python
pizzas_menos_vendidas = df_pizza.groupby('pizza_name')['quantity'].sum().sort_values(ascending=True).head(5)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=pizzas_menos_vendidas.values, y=pizzas_menos_vendidas.index, palette='Reds_r')
plt.title('Top 5 Pizzas con Menor Demanda (Baja Rotacin)', fontsize=15)
plt.xlabel('Cantidad Total de Unidades Vendidas', fontsize=12)
plt.ylabel('Nombre de la Pizza', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
pivot_pizza = df_pizza.pivot_table(index='pizza_size',
                                   columns='pizza_category',
                                   values='total_price',
                                   aggfunc='sum')
plt.figure(figsize=(10,6))
sns.heatmap(pivot_pizza, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Mapa de Calor: Ingresos por Tamao y Categora')
plt.show()
```
